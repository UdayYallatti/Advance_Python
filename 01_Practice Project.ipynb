```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = []

    def add_grade(self, grade):
        if 0 <= grade <= 100:
            self.grades.append(grade)
            print(f"Grade {grade} added.")
        else:
            print("Invalid grade! Grade must be between 0 and 100.")

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

    def display_info(self):
        print("Student Name:", self.name)
        print("Student ID:", self.student_id)
        print("Grades:", self.grades)
        print("Average Grade:", self.average_grade())
```

```python
student1 = Student("Alice", "S123")

student1.add_grade(95)
student1.add_grade(85)
student1.add_grade(110)   # Invalid
student1.add_grade(-10)   # Invalid

student1.display_info()
```

```python
class Book:
    def __init__(self, title, author, total_pages):
        self.title = title
        self.author = author
        self.total_pages = total_pages
        self.pages_read = 0

    def read_pages(self, pages):
        if pages < 0:
            print("You cannot read a negative number of pages")
        elif self.pages_read + pages > self.total_pages:
            print("You cannot read more pages than the book has")
        else:
            self.pages_read += pages
            print(f"You have read {pages} pages")

    
    def progress(self):
        return (self.pages_read/ self.total_pages)*100

    def display_info(self):
        print(f"Title of book: {self.title}")
        print(f"Author of the book: {self.author}")
        print(f"Pages read read till now: {self.pages_read} and Total pages are {self.total_pages}")
        print(f"Progress of reading: {self.progress():.2f}%")
        
```

```python
book1 = Book("Python Basics", "John Smith", 300)

book1.read_pages(50)
book1.read_pages(100)
book1.read_pages(200)  # Not allowed

book1.display_info()
```
