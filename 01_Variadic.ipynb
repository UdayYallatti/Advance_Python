```python
def add(a = int|float , b = int|float) -> int|float:
    """
    This function adds two numbers a and b
    """
    c = a + b
    return c
```

```python
def function1(*args):
    print(f"Positional Argument: {args}")
    print(f"Type : {type(args)}")
    for i in args:
        print(i)
```

```python
function1(1, "Uday", 3.5)
```

```python
a = [1,2,3, "Uday", True, False, 30.5]
print(a, type(a))
```

```python
function1(*a)
```

```python
def add_multiple(*args) -> int | float:
    s = 0
    for i in args:
        s= s + i
    return s
```

```python
add_multiple(2, 4 , 6 ,8)
```

```python
a1 = add_multiple(2, 10, 20, 50)
print(a1)
```

```python
nums = list(range(1, 1000))
print(nums)
a3 = add_multiple(*nums)
print(a3)
```

```python
cart = [
    ("Laptop", 2, 45000),
    ("Mouse", 1, 500),
    ("Keyboard", 2, 1500),
    ("Screen Gaurd", 3, 800)
]
```

```python
def billing(*args):
    total_bill = 0
    for name, qty, price in args:
        line_total = price * qty
        total_bill = total_bill + line_total
        print(f"Name : {name}, Quantity : {qty}, Price : {price}, Line total : {line_total}")
        print("="*60)
    return total_bill
```

```python
cart
```

```python
b1 = billing(*cart)
print(f"Total bill : {b1} INR")
```

```python
b2 = billing(
    ("Mobile", 2 , 25000),
    ("Cover" , 4 , 300),
    ("Screen Gaurd", 2, 400)
)
print(f"Total Bill: {b2} INR")
```

```python
def function2(**kwargs):
    print(f"Keyword Arguments: {kwargs}")
    print(f"Type : {type(kwargs)}")
    for key , value in kwargs.items():
        print(f"Key : {key}, Value: {value}")
```

```python
function2(name= "uday", poition= "Engg", salary= 100000, time_in_years= 5)
```

```python
function2(Employee="Uday" , ID=101 , Mobile= 9890987656 , Bonus= 25000)
```

```python
def function3(*args, **kwargs):
    print(f"Positional Arguments: {args}")
    print(f"Type of Positional arguments: {type(args)}")
    for i in args:
        print(i)
    print("="*60)
    print(f"Keyword Arguments: {kwargs}")
    print(f"Type of keyword arguments: {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key: {key}, Value: {value}")
        
```

```python
function3(1,2,3,"Test", "u@gmail.com",True)
```

```python
function3(1,3,5, id= 101, name= "uday", Marriage= True)
```

```python
a= [12, 15, True, "Uday", "Sayali"]
b= {"id":101,"name": "Uday", "spouse":"Sayali", "Married": True}
function3(*a, **b)
```
