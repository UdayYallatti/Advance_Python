```python
import math
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        if self.a <= 0 or self.b <= 0 or self.c <= 0:
            raise ValueError("Sides of triangle cannot be negative or zero")
        if not self.is_triangle():
            raise ValueError("Sum of any two sides should be greater than the third")
        
    def __str__(self):
        return f"Triangle(a = {self.a}, b = {self.b}, c= {self.c} )"
        
    def is_triangle(self):
        return (
                self.a + self.b > self.c and
                self.b + self.c > self.a and
                self.a + self.c > self.b
            )
        
    def perimeter(self):
        return self.a + self.b + self.c
        
    def area(self):
        s = self.perimeter()/ 2
        return math.sqrt(s*(s-self.a)*(s - self.b)*(s - self.c))
        
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def perimeter(self):
        return 2*(self.width + self.height)
    
    def area(self):
        return self.width * self.height
    
        
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def perimeter(self):
        return 2*math.pi*self.radius
    
    def area(self):
        return math.pi*(self.radius**2)
```

```python
shapes = [
    Triangle(a = 12, b = 13, c = 14),
    Circle(radius = 15),
    Rectangle(width=25, height=30),
    Circle(radius = 21),
    Triangle(a = 11, b = 12, c = 13)
]

for i in shapes:
    p = i.perimeter()
    a = i.area()
    print(type(i))
    print(f"Perimeter: {p}")
    print(f"Area: {a}")
    print("========================")
```
